```python
import pandas as pd
import os

filename='../FDDC/hetong.train'
```

```python
class hetong():
    def __init__(self,name,jiafang,yifang,xiangmu,hetong):
        self.name=name
        self.jiafang=jiafang
        self.yifang=yifang
        self.xiangmu=xiangmu
        self.hetong=hetong

    
```

```python
length,length2,ht=[],[],[]
with open(filename,'r') as fr:
    for line in fr.readlines():
        a=line.split('\t')
        length.append(len(a))
        if len(a)<8:
            a.extend(['']*(8-len(a)))
        a=a[:5]
        length2.append(len(a))
        ht.append(hetong(*a))
```

```python
import numpy as np
np_len=np.array(length2)
np_len.max(),np_len.min()
```
